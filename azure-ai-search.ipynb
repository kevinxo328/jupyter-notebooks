```python
from pydantic_settings import BaseSettings

class Env(BaseSettings):
    SEARCH_API_KEY:str
    SEARCH_INDEX:str
    SEARCH_SERVICE_NAME:str
    class Config:
        env_file = "azure-ai-search.env"
        
env = Env()
```

## Example of a full text query request
- https://learn.microsoft.com/en-us/azure/search/search-query-create?tabs=portal-text-query#example-of-a-full-text-query-request

```python
import requests

headers = {
    'api-key': env.SEARCH_API_KEY,
    'Content-Type': 'application/json'
}

url = f'https://{env.SEARCH_SERVICE_NAME}.search.windows.net/indexes/{env.SEARCH_INDEX}/docs/search?api-version=2020-06-30'

body = {
    "search": "",
    "queryType": "full",
    "searchMode": "all"
}

response = requests.post(url, headers=headers, json=body)
response.json()
```
