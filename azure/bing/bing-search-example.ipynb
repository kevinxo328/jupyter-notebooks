```python
from pydantic_settings import BaseSettings
class Settings(BaseSettings):
    BING_API_KEY: str
    BING_CUSTOM_CONFIG_ID: str = "1"
    BING_API_ENDPOINT: str = "https://api.bing.microsoft.com/"

    class Config:
        env_file = ".env"
```

## Web Search

```python
import json
from pprint import pprint
import requests 

'''
This sample uses the Bing Custom Search API to search for a query topic and get back user-controlled web page results.
Bing Custom Search API: https://docs.microsoft.com/en-us/bing/search-apis/bing-custom-search/overview 
'''
searchTerm = "microsoft"

# Add your Bing Custom Search endpoint to your environment variables.
url = Settings().BING_API_ENDPOINT + "/v7.0/custom/search?q=" + searchTerm + "&customconfig=" + Settings().BING_CUSTOM_CONFIG_ID
r = requests.get(url, headers={'Ocp-Apim-Subscription-Key': Settings().BING_API_KEY})
pprint(json.loads(r.text))
```
